```python
#这个文件是使用贝叶斯 向量机做的一个情感分析 这个文件对你这个方向没有作用了 可以不写


from sklearn.feature_extraction.text import CountVectorizer
from sklearn import svm
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn import svm
import re
import pandas as pd
import numpy as np
import jieba
```

```python
data = pd.read_csv('clean_data.csv')
comments= []
labels = []#下面这个函数 依旧是去掉中性评论
for i in range(len(data['comments'])):
    if data['labels'][i] != 1:
        comments.append(data['comments'][i])
        if data['labels'][i] == 2 :
            labels.append(data['labels'][i]-1)
        else:
            labels.append(data['labels'][i])
print(len(comments),len(labels))
'''def clean_data(a_list):   # 清理数据 清除英文等 并且分词 
    data = []
    for i in a_list:
        re.sub('\d+','',i)
        i = jieba.lcut(i.strip())
        for x in range(len(i)):
            try :
                i.remove(' ')
            except:
                break
        data.append(i)
    return data
        
data = clean_data(comments) '''
def cut(text):
    return " ".join(jieba.cut(text))
for x in range(len(comments)):
    comments[x] = cut(comments[x])#这里是将所有的评论分词
```

```python
Vectorizer = CountVectorizer(max_df = 0.8,
                            min_df = 3,
                            token_pattern =u'(?u)\\b[^\\d\\W]\\w+\\b',
                             max_features=400
                            )#这里使用的是自己搭建one-hot编码的词向量 这个是将这个模型初始化
test = pd.DataFrame(Vectorizer.fit_transform(comments).toarray(),columns = Vectorizer.get_feature_names())#将分词好的评论输入这个词向量中 

```

```python
test#这个就是那个词向量的模样 是一个稀疏矩阵
```

```python
# 加载模型 nb svm
```

```python
x_train ,x_test ,y_train ,y_test = train_test_split(comments,labels,test_size = 0.2,random_state = 0) #分割数据

x_train = Vectorizer.fit_transform(x_train)#转换训练 检测数据
x_test = Vectorizer.transform(x_test)
 
nb = MultinomialNB()  #加载模型
svm = svm.SVC(kernel = 'rbf',decision_function_shape='ovo') 
```

```python
#支持向量机

svm.fit(x_train,y_train)
svm.score(x_test,y_test)
```

```python
#贝叶斯

nb.fit(x_train,y_train)
nb.score(x_test,y_test)
```
