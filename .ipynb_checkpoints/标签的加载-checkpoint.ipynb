```python
#给爬出的数据集做清理 然后给加标签 通过百度云识别 进行标签加载
import requests
```

```python
url = 'https://ai.baidu.com/aidemo'#这里调用的一个post爬虫 使用百度情感分析接口 来给用来训练的评论做一个正负向的一个标签  这个框是一个爬虫的测试
data = { 
    'apiType': 'nlp',
'type': 'sentimentClassify',
't1':'这个东西感觉一般般'
}
```

```python
res = requests.post(url,data)
```

```python
res.json()#这里对‘这个东西感觉一般般’这句话使用百度云api进行情感分析 可看出结果是sentiment是负向评论
```

```python
import pandas as pd
import time
```

```python
com = pd.read_csv('lihuanying.csv')#读取数据
```

```python
res = res.json()
res['data']['items'][0]['sentiment']
```

```python
labels = []#对每个评论进行标签加载 需要很长时间 
comment = com['comments']
headers = {
'content-type': 'text/json'}#定义头  直有定义了头信息 才能让服务器知道 访问它的不是机器人 而是浏览器

for i in comment:#对每个评论进行遍历 然后输入百度云进行情感分析 由于数据量比较大 所以需要很久
    data = {
    'apiType': 'nlp',
    'type': 'sentimentClassify',
    't1':i
        }
    label = requests.post(url,data)
    print(label.status_code)
    print(label.text)
    label = label.json()
    label = label['data']['items'][0]['sentiment']
    labels.append(label)
    time.sleep(1)#这里每一次训练要使进程休眠一秒 不然百度云服务器会因为访问频率过高 锁ip
```

```python
print(len(comment))
print(len(labels))
```

```python
import re
comment_new = []
for com in comment:
    com = re.sub(r'[^\w\s]',' ',com)#使用正则将评论中各种标点符号去掉
    com = re.sub(r'\d+','',com)#去掉数字

    comment_new.append(com)
comment_new    
```

```python
print(len(comment_new))
```

```python
new_data = pd.DataFrame(data = {'comments':comment_new,'labels':labels})#对评论和标签整理为dataframe格式
```

```python
new_data.to_csv('',encoding='utf-8',index = None,)#保存为csv文件
```
